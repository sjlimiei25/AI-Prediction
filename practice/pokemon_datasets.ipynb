## 포켓몬 데이터셋 문제
#### Datasets Info.
https://www.kaggle.com/datasets/alopez247/pokemon/data

---
#### Columns Info.
| 컬럼명                  | 설명                                                       |
| -------------------- | -------------------------------------------------------- |
| **Number**           | 포켓몬 도감(Pokédex)의 고유 ID                                   |
| **Name**             | 포켓몬 이름                                                   |
| **Type_1**           | 포켓몬의 **주 타입(Primary Type)**                              |
| **Type_2**           | 포켓몬의 **보조 타입(Secondary Type)** — 없는 경우 빈 값               |
| **Total**            | 모든 기본 스탯(HP, Attack, Defense, Sp_Atk, Sp_Def, Speed)의 총합 |
| **HP**               | 체력(Health Points) 기본 스탯                                  |
| **Attack**           | 물리 공격(Physical Attack) 기본 스탯                             |
| **Defense**          | 물리 방어(Physical Defense) 기본 스탯                            |
| **Sp_Atk**           | 특수 공격(Special Attack) 기본 스탯                              |
| **Sp_Def**           | 특수 방어(Special Defense) 기본 스탯                             |
| **Speed**            | 스피드(Speed) 기본 스탯                                         |
| **Generation**       | 해당 포켓몬이 등장한 세대 (1~7세대 등)                                 |
| **isLegendary**      | 전설 포켓몬 여부 (True/False)                                   |
| **Color**            | 포켓몬의 포켓몬 도감상 색상                                          |
| **hasGender**        | 성별 구분이 존재하는 포켓몬인지 여부 (True/False)                        |
| **Pr_male**          | 남성 포켓몬일 확률 (Female 확률은 1 - Pr_male)                      |
| **Egg_Group_1**      | 첫 번째 알 그룹(Egg Group)                                     |
| **Egg_Group_2**      | 두 번째 알 그룹(있는 경우)                                         |
| **hasMegaEvolution** | 메가진화를 할 수 있는 포켓몬인지 여부                                    |
| **Height_m**         | 포켓몬 키(미터 단위)                                             |
| **Weight_kg**        | 포켓몬 몸무게(kg 단위)                                           |
| **Catch_Rate**       | 포획률(Catch Rate)                                          |
| **Body_Style**       | 도감에 기록된 포켓몬의 몸 형태(Body Style)                            |

> ### 다음 제시된 2가지 문제 중 하나를 선택하여 모델 구축


### [1] 포켓몬 능력치(Total) 예측 모델 만들기
포켓몬 데이터셋에는 HP, Attack, Defense 등 각 포켓몬의 세부 능력치와\
전설 여부(isLegendary), 타입(Type_1, Type_2), 세대(Generation) 등 다양한 정보가 포함되어 있다.

이번 문제의 목표는 **개별 능력치 및 포켓몬 속성 정보를 기반으로 Total(종합 능력치)을 예측하는 회귀 모델** 을 만드는 것이다.

- ### 작업 목표
  - 수치형/범주형 데이터 전처리
  - 상관관계 분석을 통해 Feature 선택
  - 학습/검증 데이터 분리
  - 회귀 모델 적용 및 성능 비교
  
- ### 모델 성능 목표
  - LinearRegression 모델
    - R² >= 0.60
    - RMSE <= 70
  - RandomForest 모델
    - R² >= 0.75
    - RMSE <= 55


### [2] 포켓몬 전설 여부 예측 모델 만들기
포켓몬 데이터셋에는 개별 능력치(HP, Attack, Defense, Sp_Atk, Sp_Def, Speed)뿐 아니라\
타입(Type_1, Type_2), 세대(Generation) 등의 범주형 정보가 포함되어 있다.

이번 과제의 목표는 **포켓몬의 능력치 및 속성 정보를 기반으로 전설 포켓몬 여부(isLegendary)를 예측하는 분류 모델** 을 구축하는 것이다.

- ### 작업 목표
  - 분류 모델 학습을 위한 타깃 변수 설정
  - 개별 스탯 및 속성 정보를 Feature로 활용하기 위해 수치형/범주형 전처리 수행
  - 학습/검증용 데이터 분리
  - 로지스틱 회귀(Logistic Regression), 랜덤 포레스트(RandomForestClassifier) 등 \
    여러 분류 모델을 적용해보고 성능 비교 분석

- ### 모델 성능 목표
  - Logistic Regression 모델
    - accuracy >= 0.90
    - f1-score >= 0.60 (전설 클래스=1 기준)
    - confusion matrix 를 통해 전설 클래스(1) 오분류 패턴 분석
  - RandomForest 모델
    - accuracy >= 0.93
    - f1-score >= 0.70
    - roc-auc >= 0.90
