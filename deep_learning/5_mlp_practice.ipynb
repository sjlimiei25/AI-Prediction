## MLP (Multi-Layer Perceptron, 다층 퍼셉트론)
- 여러 개의 퍼셉트론 층을 순차적으로 쌓아 만든 기본 신경망 구조
- 각 뉴런이 다음 층의 모든 뉴런과 연결되는 '완전 연결'방식
- 이미지처럼 2D 데이터를 1차원으로 펼쳐 입력으로 사용
  - ex. MNIST 의 경우 28x28 이미지 -> 784차원 벡터로 변환하여 입력

```python
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense

# * 데이터 준비 (mnist)
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# * 전처리 (정규화 + reshape)
x_train = x_train.reshape(-1, 28*28) / 255.0
x_test = x_test.reshape(-1, 28*28) / 255.0

print(x_train.shape)

# * 모델 정의
model = Sequential([
  Dense(128, activation='relu', input_shape=(x_train.shape[1], )),
  Dense(64, activation='relu'),
  Dense(10, activation='softmax')
])

# * 컴파일
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# * 학습
model.fit(x_train, y_train, epochs=5, validation_split=0.2)
# * validation_split : 훈련 데이터의 일부를 검증 데이터로 사용하고자 할 때 설정
#   validation_split=0.2 : 훈련 데이터의 20%를 검증용으로 분리해서 사용
#   => 과적합 여부 확인을 위해 사용

# * 평가
model.evaluate(x_test, y_test)
# [0.08022482693195343, 0.975600004196167]
# => loss, accuracy
```
